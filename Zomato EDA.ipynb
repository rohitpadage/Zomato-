```python
df_country=pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
df.columns
```

```python
final_df=pd.merge(df,df_country,on="Country Code",how='left')
final_df
```

```python
final_df.dtypes
```

```python
country_names=final_df['Country'].value_counts().index
```

```python
country_val=final_df['Country'].value_counts().values
```

```python
plt.pie(country_val[:3],labels=country_names[:3],autopct="%1.2ff%%")
```

```python
Zomato's maximum transaction are from India than USA and UK
```

```python
final_df.columns
```

```python
ratings=final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text',]).size().reset_index().rename(columns={0:'Rating count'})
```

```python
ratings
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(x='Aggregate rating',y='Rating count',data=ratings)
```

```python
sns.barplot(x='Aggregate rating',y='Rating count',hue='Rating color',data=ratings,palette=['brown','red','orange','yellow','green'])
```

```python
Observation:
    1. Not rated count is very high
    2. Maximum no. of rating are between 2.5 to 3.4
```
